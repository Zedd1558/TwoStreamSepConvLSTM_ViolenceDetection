```
# opencv might not read mpg files properly, so converting them to mp4
!pip install MoviePy
print('converting all mpg files to mp4')
import os 
import moviepy.editor as moviepy
source = '/content/violenceDetection/raw_videos/movies'
for filename in os.listdir(source):
  filepath = os.path.join(source, filename)
  if filename.endswith('.mpg'):
    clip = moviepy.VideoFileClip(filepath)
    os.remove(filepath)
    newfilepath = filepath[:-4] + '.mp4'
    clip.write_videofile(newfilepath)
print('done!') 
```
